Provide a boxplot for the MEDV variable vs the AGE variable. (Discretize the age variable into three groups of 35 years and younger, between 35 and 70 years and 70 years and older)


```python
import numpy as np
ax = sns.boxplot(x=boston_df['AGE']<35, y='MEDV', data=boston_df,color='red')
ax = sns.boxplot(x=(boston_df['AGE']>35) & (boston_df['AGE']<70), y='MEDV', data=boston_df,color='yellow')
ax = sns.boxplot(x=(boston_df['AGE']>70), y='MEDV', data=boston_df,color='blue')

ax.set_title('MEDV vs. Age')
plt.show()
```

Provide a scatter plot to show the relationship between Nitric oxide concentrations and the proportion of non-retail business acres per town. What can you say about the relationship?


```python
ax = sns.scatterplot(x='INDUS',y='NOX', data=boston_df,alpha=0.05)
plt.title('Nitric Oxide vs. non-retail business acres/town')
plt.show()
```
